```python
%matplotlib inline
import geopandas as gpd
from geopandas.tools import sjoin
import matplotlib.pyplot as plt
```

```python
#load files
parcel = gpd.read_file("data/uk_land_registry_parcel_extract.geojson")
building = gpd.read_file("data/uk_osm_building_extract.geojson")
#plot
#parcel.plot(figsize = (10,10), color = 'red', alpha = 0.5, edgecolor = 'black')
#building.plot(figsize = (10,10), alpha = 0.5, edgecolor = 'black')
fig, ax = plt.subplots(figsize = (10,10))
parcel.plot(ax = ax, edgecolor = 'black', color = 'none')
building.plot(ax = ax, alpha = 0.5)

# reduce building size
building.crs == parcel.crs
building.geometry = building.geometry.scale(xfact=0.8/1, yfact=0.8/1, zfact=1.0)
from geopandas.tools import sjoin
join_left_df = sjoin(gpd.GeoDataFrame(parcel), gpd.GeoDataFrame(building), how="left", op="intersects")

# plot reduced building size
fig, ax = plt.subplots(figsize = (10,10))
parcel.plot(ax = ax, edgecolor = 'black', color = 'none')
building.plot(ax = ax, alpha = 0.5)
```
